```python
def combine_datasets(submission_file, fname=None):
    submission_df = pd.read_csv(submission_file)
    user_df = pd.read_csv('{}/user_data.csv'.format(TRAIN_PATH))
    problem_df = pd.read_csv('{}/problem_data.csv'.format(TRAIN_PATH))
    sub_prob_df = pd.merge(submission_df, problem_df, how='left', on='problem_id')
    df = pd.merge(sub_prob_df, user_df, on='user_id', how='left')
    if fname:
        df.to_csv('data/{}'.format(fname), index=False)
    return df
```

```python
train_df = combine_datasets('data/train/train_submissions.csv', 'train.csv')
train_df.head()
```

```python
test_df = combine_datasets('data/test_submission.csv', 'test.csv')
test_df.head()
```
